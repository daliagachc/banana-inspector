```python
def open_darray( pp ):
    p = os.path.join( here_path , pp )
    ip_darray = np3.fu.open_sum(p)
    
    ip_darray = np3.fu.get_treated_da(ip_darray)
    
    
    da = np3.fu.set_secs_dim(ip_darray)
    da['tempK'] = xr.zeros_like(da['secs']) + 273
    da['presP'] = xr.zeros_like(da['secs']) + 532000
    
    return da
```
